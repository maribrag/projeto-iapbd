```python
import requests
from bs4 import BeautifulSoup

# Making a GET request
r = requests.get('https://www.ncbi.nlm.nih.gov/nuccore/AH002367.2')
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# Procurar um tag meta com um determinado atributo
lines = soup.find_all('meta', {'name':"ncbi_uidlist"} )

id = ""
url = ""
for line in lines:
	# print(line)
	# if 'name' in line.attrs:
	# 	print(line.attrs['name'])
	if 'content' in line.attrs:
		# print(line.attrs['content'])		
		id = line.attrs['content']

if id:
	url = "https://www.ncbi.nlm.nih.gov/sviewer/viewer.fcgi?id={}&db=nuccore&report=genbank&conwithfeat=on&hide-cdd=on&retmode=text&ncbi_phid=CE88F25338A215A1000000000483042A&withmarkup=on&tool=portal&log$=seqview&maxdownloadsize=1000000".format(id)

r2 = requests.get( url )

print( r2.content ) 
```

```python
fonte = open("neurospora.gb", "r")
locus = fonte.read()
fonte.close()
print(locus)
```

LINK NCBI

```python
import re
existe = re.search(r'ACCESSION\s+[^\s]+', locus)
if existe:
    # print( existe[0] )
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

DEFINITION

```python
existe = re.findall(r'DEFINITION\s+.*?(?=ACCESSION)', locus, re.DOTALL)
if existe:
    for definition in existe:
        m = re.match( r'DEFINITION\s+(.+)', definition, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

ID

```python
existe = re.findall(r'LOCUS\s + [^\s]+', locus, re.DOTALL)
if existe:
    for loc in existe:
        m = re.match( r'LOCUS\s+(.+)', loc, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

LOCUS

```python
existe = re.findall(r'LOCUS\s+.*?(?=DEFINITION)', locus, re.DOTALL)
if existe:
    for definition in existe:
        m = re.match( r'LOCUS\s+(.+)', definition, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

SOURCE

```python
existe = re.findall(r'SOURCE\s+.*?(?=ORGANISM)', locus, re.DOTALL)
if existe:
    for definition in existe:
        m = re.match( r'SOURCE\s+(.+)', definition, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

ORGANISM

```python
existe = re.findall(r'ORGANISM\s+.*?(?=\n)', locus, re.DOTALL)
if existe:
    for definition in existe:
        m = re.match( r'ORGANISM\s+(.+)', definition, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

FEATURES

```python
misc = []
existe = re.findall(r'FEATURES\s+.*?(?=ORIGIN)', locus, re.DOTALL)
if existe:
    for definition in existe:
        m = re.match( r'FEATURES\s+(.+)', definition, re.DOTALL )
        misc.append(re.sub(r'\s+', '\n', m.group(1) ) )
    for c in misc:
        print(c, "\n")
```

SEQUENCE

```python
existe = re.findall(r'ORIGIN\s+.*?(?=//)', locus, re.DOTALL)
if existe:
    for nucleo in existe:
        m = re.match( r'ORIGIN\s+(.+)', nucleo, re.DOTALL)
        x = re.match(r'site.\n(.+)', m, re.DOTALL)
        a = re.sub(r'\s+', ' ', x.group(1) )
        b = a.replace(' ','')
        num = re.sub(r'\d+','', b)
        print(num)
```
